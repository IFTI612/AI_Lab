# 🧠 Brain Hemorrhage Outcome Prediction

**Domain:** Healthcare  
**Dataset:** Brain Hemorrhage Patient Records — 3,000 samples  
**Target:** Adverse outcome (Fatal / Severe Disability) vs. Better outcome (Recovered / Rehabilitation Required)  
**Models:** Decision Tree · Naïve Bayes · K-Nearest Neighbors  
**Library:** Scikit-learn · Pandas · Matplotlib · Seaborn

## ⚙️ Section 4 — Data Preprocessing

Steps:
1. Drop irrelevant columns (`Patient ID`, `Date`, `Year`)
2. Engineer a **binary target variable** from the 4-class `Outcome`
3. Label-encode remaining categorical features
4. Verify no missing values remain


```python
# ── Step 1: Drop irrelevant columns ──────────────────────────────────────────
# 'Patient ID' → unique identifier, no predictive value
# 'Date'       → raw string; Year already captures temporal context
# 'Year'       → year of admission, not a clinical predictor
drop_cols = ["Patient ID", "Date", "Year"]
df.drop(columns=drop_cols, inplace=True)
print(f"✅ Dropped columns: {drop_cols}")
print(f"   Remaining columns: {list(df.columns)}")
```

```python
# ── Step 2: Engineer binary target ────────────────────────────────────────────
# Adverse (1)  →  Fatal  OR  Severe Disability
# Better  (0)  →  Recovered  OR  Rehabilitation Required
df["Target"] = df["Outcome"].apply(
    lambda x: 1 if x in ["Fatal", "Severe Disability"] else 0
)

print("✅ Binary target created.")
print("\nClass distribution:")
counts = df["Target"].value_counts().rename({0: "Better (0)", 1: "Adverse (1)"})
print(counts.to_string())
print(f"\nClass balance: {counts[0]/len(df)*100:.1f}% Better  /  {counts[1]/len(df)*100:.1f}% Adverse")

# Drop the original multi-class Outcome column
df.drop(columns=["Outcome"], inplace=True)
```

```python
# ── Step 3: Label-encode categorical features ─────────────────────────────────
# LabelEncoder maps each unique string to a unique integer
# This satisfies the 'numeric features only' requirement
cat_cols = df.select_dtypes(include="object").columns.tolist()
le = LabelEncoder()

for col in cat_cols:
    df[col] = le.fit_transform(df[col])
    print(f"   Encoded '{col}'  →  {df[col].nunique()} unique integer codes")

print("\n✅ All categorical columns encoded.")
```

```python
# ── Step 4: Verify cleanliness ────────────────────────────────────────────────
missing = df.isnull().sum().sum()
print(f"Missing values remaining : {missing}")
print(f"All-numeric columns      : {all(df.dtypes != 'object')}")
print(f"\nFinal dataset shape      : {df.shape}")
print("\nData types after preprocessing:")
print(df.dtypes)
```

```python
# ── Preview the preprocessed dataset ─────────────────────────────────────────
print("── First 5 rows after preprocessing ─────────────────────────────────────")
display(df.head())
```

---
## ✂️ Section 5 — Train / Test Split (80% / 20%)

- `test_size=0.20` → 600 test samples  
- `stratify=y` → preserves class ratio in both splits  
- `random_state=42` → reproducibility


```python
# ── Separate features (X) and target (y) ─────────────────────────────────────
X = df.drop(columns=["Target"])
y = df["Target"]

# ── Stratified split ──────────────────────────────────────────────────────────
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,
    random_state=42,
    stratify=y          # ensures same class proportions in train & test
)

print(f"Training set   : {X_train.shape[0]} samples  ({X_train.shape[1]} features)")
print(f"Testing  set   : {X_test.shape[0]} samples")
print(f"\nTraining class balance : {y_train.value_counts().to_dict()}")
print(f"Testing  class balance : {y_test.value_counts().to_dict()}")
```

---
## 🤖 Section 6 — Model Training

Three classifiers are initialised with carefully chosen hyperparameters
and trained on the 2,400-sample training set.


### 6.1 Decision Tree Classifier
- `criterion='gini'` — Gini impurity for split quality  
- `max_depth=5` — limits tree depth to prevent overfitting  
- Rule-based splits work naturally with encoded categorical features


```python
# ── Decision Tree ─────────────────────────────────────────────────────────────
dt = DecisionTreeClassifier(
    max_depth=5,          # prevents overfitting
    criterion="gini",     # Gini impurity
    random_state=42
)
dt.fit(X_train, y_train)
print("✅ Decision Tree trained.")
```

### 6.2 Naïve Bayes (Gaussian)
- Assumes each feature follows a **Gaussian (normal) distribution**  
- Applies Bayes' theorem with conditional feature independence  
- Fast and parameter-free; priors estimated from training data


```python
# ── Naïve Bayes ───────────────────────────────────────────────────────────────
nb = GaussianNB()         # default: priors estimated from data
nb.fit(X_train, y_train)
print("✅ Naïve Bayes trained.")
```

### 6.3 K-Nearest Neighbors (KNN)
- `n_neighbors=9` — odd k avoids tie-breaking; selected via elbow curve  
- `weights='distance'` — closer neighbours carry more influence  
- `metric='euclidean'` — standard L2 distance in feature space


```python
# ── K-Nearest Neighbors ───────────────────────────────────────────────────────
knn = KNeighborsClassifier(
    n_neighbors=9,        # odd k prevents ties; tuned via elbow curve
    weights="distance",   # weight by inverse distance
    metric="euclidean"
)
knn.fit(X_train, y_train)
print("✅ KNN trained.")
```

---
## 📈 Section 7 — Model Evaluation

Each model is evaluated on the **held-out 600-sample test set** using:

| Metric | What it measures |
|--------|-----------------|
| **Accuracy** | Overall fraction of correct predictions |
| **Precision** | Of predicted Adverse, how many truly are |
| **Recall** | Of true Adverse, how many were caught (critical in healthcare) |
| **F1-Score** | Harmonic mean of Precision and Recall |


```python
# ── Collect predictions & metrics for all three models ────────────────────────
models = {
    "Decision Tree" : dt,
    "Naïve Bayes"   : nb,
    "KNN"           : knn
}

results = {}

for name, model in models.items():
    y_pred = model.predict(X_test)
    results[name] = {
        "Accuracy"  : accuracy_score (y_test, y_pred),
        "Precision" : precision_score(y_test, y_pred, zero_division=0),
        "Recall"    : recall_score   (y_test, y_pred, zero_division=0),
        "F1"        : f1_score       (y_test, y_pred, zero_division=0),
        "CM"        : confusion_matrix(y_test, y_pred),
        "y_pred"    : y_pred
    }

print("✅ Predictions generated for all models.")
```

```python
# ── Detailed classification reports ──────────────────────────────────────────
for name, res in results.items():
    print(f"{'='*55}")
    print(f"  {name}")
    print(f"{'='*55}")
    print(f"  Accuracy  : {res['Accuracy']:.4f}")
    print(f"  Precision : {res['Precision']:.4f}")
    print(f"  Recall    : {res['Recall']:.4f}")
    print(f"  F1-Score  : {res['F1']:.4f}")
    print()
    print(classification_report(
        y_test, res["y_pred"],
        target_names=["Better (0)", "Adverse (1)"]
    ))
```

```python
# ── Summary comparison table ─────────────────────────────────────────────────
metrics_df = pd.DataFrame({
    name: {k: v for k, v in res.items() if k not in ("CM", "y_pred")}
    for name, res in results.items()
}).T

print("── Performance Summary ───────────────────────────────────────────────────")
display(metrics_df.round(4).style
        .highlight_max(axis=0, color="#d4edda")
        .set_caption("Green = best value per metric"))
```

---
## 🔧 Section 9 — Hyperparameter Tuning: KNN Elbow Curve

Testing odd values of k from 1 to 25 to identify the optimal number of neighbours.


```python
# ── KNN elbow curve — accuracy vs. k ─────────────────────────────────────────
k_vals, acc_vals = [], []

for k in range(1, 26, 2):      # odd k values only
    m = KNeighborsClassifier(n_neighbors=k, weights="distance")
    m.fit(X_train, y_train)
    acc_vals.append(accuracy_score(y_test, m.predict(X_test)))
    k_vals.append(k)

best_k   = k_vals[np.argmax(acc_vals)]
best_acc = max(acc_vals)

fig, ax = plt.subplots(figsize=(10, 4.5))
ax.plot(k_vals, acc_vals, marker="o", color="#55A868",
        linewidth=2, markersize=7, markerfacecolor="white", markeredgewidth=2)
ax.axvline(x=9, color="#DD8452", linestyle="--", linewidth=1.8, label="k=9 chosen")
ax.scatter([best_k], [best_acc], color="#E74C3C", s=100, zorder=5,
           label=f"Best k={best_k} (acc={best_acc:.4f})")
ax.set_xlabel("Number of Neighbours (k)", fontsize=12)
ax.set_ylabel("Test Accuracy", fontsize=12)
ax.set_title("KNN — Accuracy vs. k (Elbow Curve)", fontsize=13, fontweight="bold")
ax.legend()
ax.grid(alpha=0.3, linestyle="--")
ax.spines[["top", "right"]].set_visible(False)

plt.tight_layout()
plt.show()

print(f"Best k: {best_k}  |  Accuracy: {best_acc:.4f}")
```

## 📝 Section 11 — Conclusion & Analysis

### Results Summary

| Model | Accuracy | Precision | Recall | F1-Score |
|-------|----------|-----------|--------|----------|
| **Decision Tree** ✅ | **0.6333** | **0.6077** | **0.8123** | **0.6953** |
| Naïve Bayes | 0.5783 | 0.5645 | 0.7929 | 0.6595 |
| KNN (k=9)   | 0.4983 | 0.5120 | 0.5534 | 0.5319 |

---

### 🥇 Decision Tree — Best Model
- Achieves the highest scores across **all four metrics**
- **Recall of 0.8123** — catches 81% of truly adverse patients (critical in clinical settings)
- Tree-based split conditions (e.g., *Hemorrhage Type = ICH AND Age > 70 → Adverse*) naturally align with encoded categorical feature structure
- `max_depth=5` provides a good bias-variance trade-off

### 🥈 Naïve Bayes — Competitive
- Reasonable performance despite its **independence assumption**
- Age, Hemorrhage Type, and Symptoms are correlated in practice — violating the Gaussian NB assumption
- Strong recall (0.7929) but lower precision → more false positives

### 🥉 KNN — Weakest
- Near-random accuracy (0.4983) due to the **curse of dimensionality**
- `Symptoms` has 341 encoded integer codes — Euclidean distance becomes semantically meaningless
- No feature scaling applied; the large numerical range of `Symptoms` dominates the distance calculation

---

### 🏥 Clinical Perspective
In brain hemorrhage triage, **recall is the most important metric** — missing a fatal case (false negative) 
is far more costly than a false alarm. The Decision Tree's recall of **0.8123** makes it the safest choice.

---

### 🚀 Potential Improvements
1. **Feature scaling** (StandardScaler) before KNN to normalise the `Symptoms` range
2. **One-hot encoding** instead of label encoding to remove artificial ordinality
3. **Random Forest** to extend Decision Tree gains while reducing variance
4. **Threshold tuning** on classifier probabilities to further optimise recall
5. **SMOTE** oversampling if class imbalance becomes a concern on different splits


```python
# ── Final summary printout ────────────────────────────────────────────────────
best_model = metrics_df["F1"].idxmax()
best_f1    = metrics_df.loc[best_model, "F1"]
best_rec   = metrics_df.loc[best_model, "Recall"]

print("=" * 55)
print("  FINAL RESULTS SUMMARY")
print("=" * 55)
print(metrics_df.round(4).to_string())
print()
print(f"🏆  Best model by F1-Score : {best_model}")
print(f"    F1-Score               : {best_f1:.4f}")
print(f"    Recall                 : {best_rec:.4f}")
print()
print("✅  Notebook execution complete.")
```
